```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_csv("out_train.csv", dtype={'Class': 'category'})
test = pd.read_csv("out_test.csv", dtype={'Class': 'category'})
```

```python
train.describe().T
```

```python
train_x, train_y = train.iloc[:,2:40], train.iloc[:,40]
test_x, test_y = test.iloc[:,2:40], test.iloc[:,40]
```

```python
norm_m = train_x.mean().to_numpy()
norm_s = train_x.std().to_numpy()
def normalize(x: np.ndarray | pd.DataFrame) -> np.ndarray:
    return ((x if isinstance(x, np.ndarray) else x.to_numpy()) - norm_m) / norm_s
```

```python
train_nx = normalize(train_x)
```

```python
rfc = RandomForestClassifier(n_estimators=300, random_state=1)
rfc.fit(train_nx, train_y)
predicted_y = rfc.predict(normalize(test_x))
sum(predicted_y == test_y) / len(test_y)
```

```python
svc = svm.LinearSVC(max_iter=100_000, random_state=1)
svc.fit(train_nx, train_y)
predicted_y = svc.predict(normalize(test_x))
sum(predicted_y == test_y) / len(test_y)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4).fit(train_x)
kmeans.labels_
```

```python
pd.crosstab(kmeans.labels_, train_y.cat.codes)
```
