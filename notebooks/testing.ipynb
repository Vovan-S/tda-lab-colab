```python
import os 

N_MFCC = 10
N_CONT = 7
FILTER_THRESHOLD = 0.4
def extract_features(directory: str, classes: str, to: str):
    files = os.listdir(directory)
    n_files = len(files)
    print(f"Reading {n_files} files from {directory}")
    data = []
    for i, file in enumerate(files):
        print(f"\r[ {' ' * (len(str(n_files)) - len(str(i+1)))}{i + 1} / {n_files} ] {file}", end='')
        y, sr = librosa.load(os.path.join(directory, file))
        yf = y[y > np.mean(np.abs(y)) * FILTER_THRESHOLD]
        y_h, y_p = librosa.effects.hpss(yf)
        mfccs = librosa.feature.mfcc(y=y_h, sr=sr, n_mfcc=N_MFCC)
        contrast = librosa.feature.spectral_contrast(y=y_h, sr=sr)
        data.append([
            file, 
            np.mean(np.abs(y_h)),
            np.std(np.abs(y_h)),
            np.mean(np.abs(y_p)),
            np.std(np.abs(y_p)),
            *np.mean(mfccs, 1),
            *np.std(mfccs, 1),
            *np.mean(contrast, 1),
            *np.std(contrast, 1),
        ])
    print("\nSuccessfully read all the files")
    df = pd.DataFrame(data=data, columns=[
        "Filename",
        "harmonic_mean",
        "harmonic_std",
        "percussive_mean",
        "percussive_std",
        *(f"mfcc{i}_mean" for i in range(N_MFCC)),
        *(f"mfcc{i}_std" for i in range(N_MFCC)),
        *(f"cont{i}_mean" for i in range(N_CONT)),
        *(f"cont{i}_std" for i in range(N_CONT)),
    ])
    classes = pd.read_csv(classes)
    df2 = pd.merge(df, classes, left_on="Filename", right_on="FileName")
    df2.pop("FileName")
    print(f"Writing to {to}")
    df2.to_csv(to, index_label="N")
```

```python
def get_chunks(
    y: np.ndarray, 
    chunk_size: int, 
    number_of_chunks: int
)-> List[Tuple[int, np.ndarray]]:    
    y_h, y_p = librosa.effects.hpss(y)
    y_p = np.abs(y_p)
    n = len(y)
    result = []
    while len(result) < number_of_chunks:
        i_max = np.argmax(y_p)
        if i_max + chunk_size > n:
            y_p[i_max:] = 0
            continue
        result.append((i_max, y_h[i_max: i_max + chunk_size]))
        y_p[i_max: i_max + chunk_size] = 0
    return result
```

```python
import h5py
```

```python
def export_chunks(
    directory: str, 
    classes: str, 
    to: str, 
    number_of_chunks: int,
    chunk_size: int):
    files = os.listdir(directory)
    n_files = len(files)
    print(f"Reading {n_files} files from {directory}")
    
    chunks = np.ndarray((n_files, number_of_chunks, chunk_size), dtype=float)
    starts = np.ndarray((n_files, number_of_chunks + 1), dtype=int)
    class_labels = []

    classes_df = pd.read_csv(classes)

    for i, file in enumerate(files):
        print(f"\r[ {' ' * (len(str(n_files)) - len(str(i+1)))}{i + 1} / {n_files} ] {file}", end='')
        y, _ = librosa.load(os.path.join(directory, file))
        ch = get_chunks(y, chunk_size, number_of_chunks)
        starts[i, 0] = len(y)
        for j, (start, c) in enumerate(ch):
            starts[i, j + 1] = start
            chunks[i, j, :] = c
        cl = classes_df[classes_df["FileName"] == file]["Class"].iloc[0]
        class_labels.append(cl)
    print()
    print(f"\nSuccessfully read {n_files} files")
    print(f"Writing to {to}")
    with h5py.File(to, mode='w') as h5file:
        h5file.create_dataset("chunks",    data=chunks)
        h5file.create_dataset("filenames", data=np.array(files, dtype='S'))
        h5file.create_dataset("starts",    data=starts)
        h5file.create_dataset("classes",   data=np.array(class_labels, dtype='S'))
    print(f"File {to} written")
```

```python
export_chunks(
    "../files/original/Test_submission/", 
    "../files/Metadata_Test.csv", 
    "../files/test.h5",
    5,
    int(sr * 0.2))
```

```python
export_chunks(
    "../files/original/Train_submission/", 
    "../files/Train_metadata_fixed.csv", 
    "../files/train.h5",
    5,
    int(sr * 0.2))
```
